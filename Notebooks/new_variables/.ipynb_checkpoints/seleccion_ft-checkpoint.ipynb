```python
import numpy as np
import pandas as pd
from functools import reduce
```

Cargamos importancias guardadas

```python
ft_imp1 = pd.read_csv('../../importances/feature_importances_lgbc_model_9.csv')
ft_imp2 = pd.read_csv('../../importances/feature_importances_lgbc_model_10.csv')
ft_imp3 = pd.read_csv('../../importances/feature_importances_lgbc_model_11.csv')
ft_imp4 = pd.read_csv('../../importances/feature_importances_lgbc_model_12.csv')
```

Normalizamos las importancias de 0 a 1 diviendo por el maximo

```python
ft_imp1['importance'] = ft_imp1['importance']/ft_imp1['importance'].max()
ft_imp2['importance'] = ft_imp2['importance']/ft_imp2['importance'].max()
ft_imp3['importance'] = ft_imp3['importance']/ft_imp3['importance'].max()
ft_imp4['importance'] = ft_imp4['importance']/ft_imp4['importance'].max()
```

```python
ft_imp1.head()
```

Mergeamos los 4 DFs en 1 solo DF

```python
df1 = ft_imp1.merge(ft_imp2, on='feature', how='inner', suffixes=('_1', '_2'))
```

```python
df2 = df1.merge(ft_imp3, on='feature', how='inner', suffixes=('_1', '_2'))
```

```python
df3 = df2.merge(ft_imp4, on='feature', how='inner', suffixes=('_3', '_4'))
```

Calculamos importance total y volvemos a normalizar

```python
df_imp = df3
df_imp['total_importance'] = df_imp['importance_1']+\
df_imp['importance_2']+\
df_imp['importance_3']+\
df_imp['importance_4']
```

```python
df_imp['total_importance'] = df_imp['total_importance']/df_imp['total_importance'].max()
```

```python
df_imp_sorted = df_imp.sort_values('total_importance', ascending=False)
```

```python
df_imp_sorted
```

Seleccionamos las 40 mas importantes por según los 4 modelos previamente generados

```python
ft_sel = df_imp_sorted['feature'][:40]
```
